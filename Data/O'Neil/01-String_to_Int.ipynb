```python
import pandas as pd
import numpy as np
```

```python
ONEIL = pd.read_csv('Synergy_ONeil.csv')
```

```python
drug_name = np.concatenate((ONEIL['drug_row'].unique(),ONEIL['drug_col'].unique()))
drug_name = set(drug_name)
Drug = pd.read_csv("Drug_MorganFingerprint.csv")
Drug = Drug.drop(columns=Drug.columns[0])
Drug = Drug.loc[(Drug['Unnamed: 0'].isin(drug_name))]
Drug.head()
```

```python
cell_name = ONEIL['DepMap_ID'].unique()
cell_name = set(cell_name)
Cell = pd.read_csv("Cell_Expression.csv")
Cell = Cell.drop(labels=['Unnamed: 0.1'],axis=1)
Cell = Cell.loc[(Cell['Unnamed: 0'].isin(cell_name))]
Cell.head()
```

```python
synergy_data = ONEIL[['drug_row','drug_col','DepMap_ID','synergy_loewe']].copy()
synergy_data = synergy_data.loc[(synergy_data['DepMap_ID'].isin(Cell['Unnamed: 0']))]
```

```python
Drug_map = dict((v, i) for i,v in enumerate(Drug['Unnamed: 0']))
Drug_id = Drug.replace({'Unnamed: 0':Drug_map})
Drug_id.rename(columns={'Unnamed: 0':'id'}, inplace = True)
Drug_id.set_index('id', inplace=True)
Drug_id.to_csv('./Preprocessed/cls/Drug_use.csv')
Drug_id.to_csv('./Preprocessed/reg/Drug_use.csv')
np.save('./Preprocessed/cls/Drug_map.npy', Drug_map)
np.save('./Preprocessed/reg/Drug_map.npy', Drug_map)
Drug_id.head()
```

```python
Cell_map = dict((v, i) for i,v in enumerate(Cell['Unnamed: 0']))
Cell_id = Cell.replace({'Unnamed: 0':Cell_map})
Cell_id.rename(columns={'Unnamed: 0':'id'}, inplace = True)
Cell_id.set_index('id', inplace=True)
Cell_id.to_csv('./Preprocessed/cls/Cell_use.csv')
Cell_id.to_csv('./Preprocessed/reg/Cell_use.csv')
np.save('./Preprocessed/cls/Cell_map.npy', Cell_map)
np.save('./Preprocessed/reg/Cell_map.npy', Cell_map)
Cell_id.head()
```

```python
synergy_data_id = synergy_data.replace({'DepMap_ID':Cell_map})
synergy_data_id = synergy_data_id.replace({'drug_row':Drug_map})
synergy_data_id = synergy_data_id.replace({'drug_col':Drug_map})
synergy_data_id
```

```python
synergy_data_id.to_csv('Synergy_ONeil_int.csv')
```
